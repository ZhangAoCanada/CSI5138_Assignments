```python
import numpy, math
import matplotlib.pyplot as plt

N = 200
theta = 1
def getData():
    
    x = numpy.random.uniform(0,1,N)

    def fun(xi):
        yi = math.cos(2*math.pi*xi)+(1/math.sqrt(2*math.pi*math.pow(theta,2)))*math.exp(-1*math.pow(xi,2)/(2*math.pow(theta,2)))
        return yi

    y=[]
    for xi in x:
        y.append(fun(xi))
    return x,y

x,y=getData()
plt.plot(x,y,'ro')
```

```python
import random
d = 20
a = []
for i in range(0,d):
    a.append(random.random())
```

```python
def predict_function(d,a,xi):
    yi_hat = 0
    for i in range(d):
        yi_hat = yi_hat + a[i]*math.pow(xi,i)
    return yi_hat

y_hat = []
for xi in x:
    y_hat.append(predict_function(d,a,xi))
    
plt.plot(x,y_hat,'ro')
```

```python
def getMSE(y,y_hat):
    mse = 0
    for i in range(len(y)):
        mse = mse + math.pow((y[i]-y_hat[i]),2)
        mse = mse/N
    return mse

getMSE(y,y_hat)
learning_rate = 0.00001
```

```python
def fitData_SGD(y,y_hat,x,a,running_threshold):
    k = 0
    number = 0
    while 1:
        a_old = a
        for i in range(1,d):
            derivative = 2*(y[k]-y_hat[k])*math.pow(x[i],i)
            a[i] = a[i] + learning_rate*derivative
        k = k+1
        number = number+1
        if(a==a_old):
            break
        if(k==N):
            k=0
        if(number == running_threshold):
            break
    return a
```

```python
def fitData_GD(y,y_hat,x,a):
    for i in range(0,d):
        derivative = 0
        for k in range(N):
            derivative = derivative + 2*(y[k]-y_hat[k])*math.pow(x[i],i)
        derivative = derivative/N
        a[i] = a[i] + learning_rate*derivative
    return a
plt.plot(x,y,'go')
learning_threshold = 10000000
for time in range(learning_threshold):
    a_old = a
#     print("a_old",sum(a_old))
    a = fitData_GD(y,y_hat,x,a)
#     print("a",sum(a))
#     a = fitData_SGD(y,y_hat,x,a,1000)
#     print("y_hat_init",sum(y_hat))
    for i in range(N):
        y_hat[i]=predict_function(d,a,x[i])
#     print("y_hat",sum(y_hat))
    if time%1000==0:
        print(getMSE(y,y_hat))
        plt.plot(x,y_hat,'ro')
    
#     if(a_old == a):
#         print("break")
#         break;
print(getMSE(y,y_hat))
plt.plot(x,y_hat,'bo')
    
```
